- Создадим выборку с 200 новыми наблюдениями, используя random_state = 21
- Выборка не должна содержать новых комбинаций номера автомобиля, марки и модели, чтобы весь набор данных оставался согласованным в этом отношении
- Нет ограничений на столбцы refund и fines. Случайным образом выберем значение из этих столбцов и применим его к любому номеру автомобиля
- Объединим выборку с исходным датафреймом, чтобы создать новый датафрейм concat_rows

```python
unique_combinations = df[['CarNumber', 'Make', 'Model']].drop_duplicates()
np.random.seed(21)
sampled_combinations = unique_combinations.sample(n=200, random_state=21)
refund_samples = np.random.choice([True, False], size=200)
fines_samples = np.random.uniform(df['Fines'].min(), df['Fines'].max(), size=200)

new_samples = sampled_combinations.copy()

new_samples['Refund'] = refund_samples
new_samples['Fines'] = fines_samples

concat_rows = pd.concat([df, new_samples], ignore_index=True)
concat_rows
```

- Создадим серию с названием Year со случайными целыми числами от 1980 до 2019
- Используйте np.random.seed(21) перед генерацией лет
- Объедините серию с датафреймом и назовите его `fines`

```python
serias = pd.Series(np.random.randint(1980, 2020, size=len(concat_rows)))
concat_rows['Year'] = serias
concat_rows
```

Создадим новый датафрейм с номерами автомобилей и их владельцами. Получим самые популярные фамилии в США.

```python
df2 = pd.read_json('../../datasets/surname.json')
df2.value_counts()
```

Создадим новую серию с фамилиями. Они не должны содержать специальных символов, таких как запятые или скобки, а их количество должно = количеству уникальных номеров автомобилей в выборке.

```python
np.random.seed(21)

def clean_surname(surname):
    replaced_surname = ''
    for letter in str(surname):
        if letter not in ".',:;!?/\" ":
            replaced_surname += letter
    return replaced_surname

cleaned_surnames = df2.iloc[:, 0].apply(clean_surname) 
selected_surnames = cleaned_surnames.sample(n=concat_rows['CarNumber'].nunique(), replace=True, random_state=21)
surnames_series = pd.Series(selected_surnames.values, name='Surname')
surnames_series
```

Создадим датафрейм owners с двумя столбцами: CarNumber и SURNAME

```python
owners = pd.DataFrame({'CarNumber': df['CarNumber'].drop_duplicates().values, 'SURNAME': surnames_series})
owners
```

Добавим пять наблюдений в датафрейм fines

```python
fines = concat_rows.copy()
new_data = pd.DataFrame({'CarNumber': ['A111AA77RUS', 'B222BB77RUS', 'C333CC77RUS', 'D444DD77RUS', 'E555EE77RUS'],'Make': ['Tesla', 'Porsche', 'Lexus', 'Volvo', 'Mazda'], 'Model': ['Model 3', '911', 'RX350', 'XC90', 'CX-5'], 'Make_n_model': ['Tesla Model 3', 'Porsche 911', 'Lexus RX350', 'Volvo XC90', 'Mazda CX-5'], 'Fines': [2500.0, 1800.0, 3200.0, 1500.0, 2700.0], 'Refund': [False, True, False, True, False], 'Year': [2020, 2021, 2019, 2022, 2018]})
fines = pd.concat([fines, new_data], ignore_index=True)
fines
```

Удалим последние 20 наблюдений из датафрейма owners и добавим три новых наблюдения, которые не совпадают с добавленными в датафрейм fines.

```python
owners = owners.iloc[:-20] 
new_data = pd.DataFrame([['M123MP77RUS', 'KOVALEV'],['N456NO77RUS', 'ORLOVA'],['P789PR77RUS', 'ROMANOV']], columns=['CarNumber', 'SURNAME'])
owners = pd.concat([owners, new_data], ignore_index=True)
owners
```

Объединим два датафрейма:
- Новый датафрейм должен содержать только номера автомобилей, которые существуют в обоих датафреймах.
- Новый датафрейм должен содержать все номера автомобилей из обоих датафреймов.
- Новый датафрейм должен содержать только номера автомобилей из датафрейма `fines`.
- Новый датафрейм должен содержать только номера автомобилей из датафрейма `owners`.

```python
inner_merged = pd.merge(fines, owners, on='CarNumber', how='inner')
full_merged = pd.merge(fines, owners, on='CarNumber', how='outer')
left_merged = pd.merge(fines, owners, on='CarNumber', how='left')
right_merged = pd.merge(fines, owners, on='CarNumber', how='right')
```

Создадим сводную таблицу из датафрейма fines, в которой значения - это суммы штрафов

```python
pivot_table = fines.pivot_table(values='Fines', index='Year', columns='Make', aggfunc='sum')
pivot_table
```
