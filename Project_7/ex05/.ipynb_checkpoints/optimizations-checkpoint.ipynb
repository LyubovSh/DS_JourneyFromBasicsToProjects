### Во всех следующих подзадачах вычислим `fines/refund*year` для каждой строки. 

1) Напишем функцию, которая перебирает датафрейм с помощью `for i in range(0, len(df))`, `iloc` и `append()` в список

```python
%%timeit
def option_1(df):
    result = []  
    for i in range(len(df)):
        fine = df.iloc[i]['Fines']
        refund = df.iloc[i]['Refund']
        year = df.iloc[i]['Year']
        if (refund * year) != 0:
            result.append(fine / (refund * year))
        else:
            result.append(0)
    return result

df['option_1'] = option_1(df)
```

```python
df.head()
```

2) Сделаем то же самое с помощью iterrows()

```python
%%timeit
def option_2(df):
    result = []
    for index, row in df.iterrows():
        fine = row['Fines']
        refund = row['Refund']
        year = row['Year']
        if (refund * year) != 0:
            result.append(fine / (refund * year))
        else:
            result.append(0)
    return result

df['option_2'] = option_2(df)
```

```python
df.head()
```

Сделаем то же самое с помощью `apply()` и lambda-функции

```python
%%timeit
df['option_3'] = df.apply(lambda x: (x['Fines'] / (x['Refund'] * x['Year'])) if (x['Refund'] * x['Year']) != 0 else 0, axis=1)
```
